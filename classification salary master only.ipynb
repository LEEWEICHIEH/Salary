```python
import matplotlib.pyplot as plt
import seaborn as sns

# 計算每個 classification 的平均薪資
mean_salaries = master_df.groupby('classification')['salary'].mean()

# 繪圖
plt.figure(figsize=(8, 6))
sns.scatterplot(data=master_df, x='classification', y='salary', hue='classification', palette='deep')

# 加上平均線
for cls, mean_salary in mean_salaries.items():
    plt.hlines(y=mean_salary, xmin=int(cls)-0.2, xmax=int(cls)+0.2,
               colors='black', linestyles='dashed', label=f'Mean - Class {cls}')

plt.xlabel("Classification")
plt.ylabel("Salary")
plt.title("Scatter Plot of Classification vs. Salary with Class-wise Means (Master Only)")
plt.legend(title="Classification and Means")
plt.grid(True)
plt.tight_layout()
plt.show()
```
