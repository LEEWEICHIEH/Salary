```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 假設你已經讀入了 Excel 資料：
df = pd.read_excel("C:/Users/william/Desktop/薪水.xlsx", sheet_name="工作表1")

# 將學歷轉為類別變數，如果還沒做的話
df['degree'] = df['degree'].astype(str)
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')

# 畫箱型圖
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")
ax = sns.boxplot(x='degree',y='salary', data=df, showfliers=False)
ax.set_title('Salary Distribution by Degree', fontsize=16)
ax.set_xlabel('Degree', fontsize=12)
ax.set_ylabel('Salary', fontsize=12)
plt.tight_layout()
plt.show()
```
